```python
from ultralytics import YOLO
```

```python
import os, random, shutil

img_dir = "data/images"
lbl_dir = "data/labels"

images = [f for f in os.listdir(img_dir) if f.endswith((".png", ".jpg", ".jpeg"))]
random.shuffle(images)

split = int(0.9 * len(images))

for i, img in enumerate(images):
    name = os.path.splitext(img)[0]
    label = name + ".txt"

    if not os.path.exists(f"{lbl_dir}/{label}"):
        continue  # skip if no label

    folder = "train" if i < split else "val"

    os.makedirs(f"{img_dir}/{folder}", exist_ok=True)
    os.makedirs(f"{lbl_dir}/{folder}", exist_ok=True)

    shutil.move(f"{img_dir}/{img}", f"{img_dir}/{folder}/{img}")
    shutil.move(f"{lbl_dir}/{label}", f"{lbl_dir}/{folder}/{label}")
```

```python
model = YOLO('yolo26s.pt')
```

```python
model.train(
    data="data/data.yaml",
    epochs=120,
    imgsz=640,
    batch=16,
    device=0
)
```
